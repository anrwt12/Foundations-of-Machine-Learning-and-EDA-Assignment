***Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?***

ANSWER

***Underfitting:***

 Model too simple; cannot capture underlying patterns. Symptoms: high training error and high validation/test error. Causes: too simple model (e.g., linear for non-linear data), too few features, high bias.

***Overfitting:***

 Model too complex; learns noise in training data. Symptoms: low training error but high validation/test error. Causes: overly complex model (deep tree, high-degree polynomial), small dataset, noise.

***Bias–Variance Tradeoff:***



  -> Bias = error from wrong assumptions (underfit).

->Variance = error from sensitivity to training data (overfit).
Goal: find model complexity that balances both.

->Detection: Compare training vs validation (or cross-validation) errors. Use learning curves (plot error vs dataset size). Large gap (train much better than val) → overfitting. Both errors high → underfitting.

***Prevention / Remedies:***

->Cross-validation (k-fold) to estimate generalization.

->Regularization: L1/L2 penalties, dropout (DL).

->Simplify model: reduce parameters, prune trees.

->More data (or data augmentation).

->Early stopping during training (monitor val loss).

->Ensembling (bagging reduces variance; boosting trades bias/variance).

***Question 3***

***How would you handle missing values in a dataset? Explain at least three methods with examples.***

Answer:

***Deletion (Listwise / Row-wise or Column-wise):***

-> When: Missing is small fraction and data is MCAR (missing completely at random).

-> How: Drop rows (df.dropna()) or drop columns with many missing values.

-> Pros/Cons: Simple; may lose information and bias results if not MCAR.

***Imputation with Summary Statistics (mean/median/mode):***

-> Numeric: Replace missing numeric values with mean (sensitive to outliers)    or  median (robust). Example: df['Age'].fillna(df['Age'].median()).

-> Categorical: Replace with mode (most frequent). Example: df['Gender'].fillna(df['Gender'].mode()[0]).

-> pros/Cons: Simple and fast; underestimates variance; can introduce bias.

***Predictive Modeling (Model-based Imputation):***

-> How: Use other features to predict missing values (e.g., regression for numeric, classification for categorical). Example: train a regression model to predict missing Age from Pclass, Sex, Fare.

-> Tools: IterativeImputer (sklearn), KNNImputer (scikit-learn), MICE.

-> Pros/Cons: Often more accurate; more complex and risk of leakage if not done properly (must fit only on train in cross-validation).

***Question 4***

***What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical).***

***Answer:***

Imbalanced dataset: One or more classes have far fewer examples than others (e.g., fraud detection where fraud << non-fraud). This causes classifiers to be biased towards majority class.

***Techniques:***

***Resampling:***

  ->Oversampling (e.g., replicate minority examples or synthesize new ones).

 ->SMOTE (Synthetic Minority Over-sampling Technique): Generates synthetic minority samples by interpolating between neighbors. Practical: use imblearn.over_sampling.SMOTE().

  ->Undersampling (reduce majority class): RandomUnderSampler or cluster-based undersampling.

  ->Pros/Cons: Oversampling increases training size and may overfit;     undersampling may lose information.

***Algorithm-level approaches / Class weights:***

->Use class weights in loss functions to penalize mistakes on minority class more heavily (e.g., class_weight='balanced' in scikit-learn, or pos_weight in XGBoost).


->Ensemble methods: BalancedRandomForest, EasyEnsemble.

->Evaluation: Use precision, recall, F1, ROC-AUC, PR curve (not accuracy). Use stratified CV.

***Question 5***

***Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.***

***Answer:***

->Why scale? Many ML algorithms depend on the scale of features: distance-based algorithms (KNN, KMeans, SVM with RBF) and gradient-based optimization (logistic regression, neural nets) converge faster and perform better when features are scaled.

***->Min–Max Scaling (Normalization):***

When to use: When you need bounded features (e.g., image pixel values). Sensitive to outliers (min/max affected).

Standardization (Z-score):

When to use: Algorithms that assume standardized data; less affected by outliers than min-max (but still influenced). Useful when features are roughly Gaussian.

Comparison: Min–max preserves original distribution shape but rescales range; standardization centers and scales by variance. Choose based on algorithm and data distribution.

***Question 6***

***Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?***

***Answer:***

***Label Encoding:*** Map each category to an integer (e.g., {'red':0, 'green':1, 'blue':2}).

***Use when:*** Categorical variable is ordinal (order matters) — e.g., low < medium < high. Or when tree-based models are used and there are many categories (but careful).

***Caveat: ***For nominal categories (no order), integer labels may create spurious ordinal relationships for linear/distance models.

***One-Hot Encoding:*** Create binary column per category (e.g., is_red, is_green, is_blue).

***Use when: ***Categorical variable is nominal and number of categories is not too large. Good for linear models, distance-based methods.

***Caveat:*** Can cause high dimensionality if many categories (use target encoding or embeddings instead).

```python
"""Question 8: Titanic Dataset
a) Compare the survival rates based on passenger class (Pclass). Which class had the highest
survival rate, and why do you think that happened?
b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and
adults (Age ≥ 18). Did children have a better chance of survival?
Dataset: https://github.com/MasteriNeuron/datasets.git
(Include your Python code and output in the code box below.)
"""

import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/titanic.csv"
t = pd.read_csv(url)

# Ensure proper types
t['Survived'] = t['Survived'].astype(int)
# Pclass survival rates
pclass_stats = t.groupby('Pclass')['Survived'].agg(['mean','count']).rename(columns={'mean':'survival_rate'})
pclass_stats['survival_rate_pct'] = pclass_stats['survival_rate']*100
print("Survival rate by Pclass:")
print(pclass_stats)

# Plot bar chart
pclass_stats['survival_rate_pct'].plot(kind='bar', title='Survival rate by Pclass')
plt.ylabel('Survival rate (%)')
plt.show()

# Age effect: create children/adults
t['Age'] = pd.to_numeric(t['Age'], errors='coerce')
t['Age_group'] = t['Age'].apply(lambda x: 'Child' if pd.notnull(x) and x < 18 else 'Adult')
age_stats = t.groupby('Age_group')['Survived'].agg(['mean','count']).rename(columns={'mean':'survival_rate'})
age_stats['survival_rate_pct'] = age_stats['survival_rate']*100
print("\nSurvival rate by Age group:")
print(age_stats)

# Cross Pclass & Age_group
cross = t.groupby(['Pclass', 'Age_group'])['Survived'].mean().unstack()
print("\nSurvival rate by Pclass and Age group (proportions):")
print((cross*100).round(2))
```

```python
"""Question 9: Flight Price Prediction Dataset
a) How do flight prices vary with the days left until departure? Identify any exponential price
surges and recommend the best booking window.
b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are
consistently cheaper/premium, and why?
Dataset: https://github.com/MasteriNeuron/datasets.git
(Include your Python code and output in the code box below.)
"""
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/flight_price.csv"
df = pd.read_csv(url)

print("Columns:", df.columns.tolist())

# Convert numeric columns
df['days_left'] = pd.to_numeric(df['days_left'], errors='coerce')
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# ---- Part A: Price vs Days Left ----
agg = df.dropna(subset=['days_left', 'price']).groupby('days_left')['price'].median().reset_index()

plt.figure(figsize=(10,5))
plt.plot(agg['days_left'], agg['price'])
plt.xlabel('Days Left to Departure')
plt.ylabel('Median Price')
plt.title('Median Flight Price vs Days Left to Departure')
plt.grid(True)
plt.show()

# Detect surges
agg['pct_change'] = agg['price'].pct_change() * 100
surges = agg[agg['pct_change'] > 30]
print("Detected price surges (>30%):")
print(surges)

# ---- Part B: Airline Comparison for Delhi–Mumbai ----
route_filter = (df['source_city'].str.contains("Delhi", case=False)) & \
               (df['destination_city'].str.contains("Mumbai", case=False))

route_df = df[route_filter]

airline_stats = route_df.groupby('airline')['price'].agg(['mean','median','count']).sort_values('median')
print("\nAirline Price Comparison (Delhi–Mumbai):")
print(airline_stats)

# Boxplot
top_airlines = airline_stats[airline_stats['count'] >= 20].index
plt.figure(figsize=(12,6))
route_df[route_df['airline'].isin(top_airlines)].boxplot(column='price', by='airline', rot=45)
plt.title("Price Distribution by Airline (Delhi–Mumbai)")
plt.suptitle("")
plt.ylabel("Price")
plt.show()
```

```python
"""Question 10: HR Analytics Dataset
a). What factors most strongly correlate with employee attrition? Use visualizations to show key
drivers (e.g., satisfaction, overtime, salary).
b). Are employees with more projects more likely to leave?
Dataset: hr_analytics
"""
# Q10_hr_analytics.py
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/hr_analytics.csv"
hr = pd.read_csv(url)

# Inspect columns
print("Columns:", hr.columns.tolist())
print("Sample rows:")
print(hr.head())

# Ensure attrition column binary (e.g., 'Attrition' -> 1/0 or 'Yes'/'No')
if 'Attrition' in hr.columns:
    if hr['Attrition'].dtype == object:
        hr['Attrition_flag'] = hr['Attrition'].map({'Yes':1, 'No':0})
    else:
        hr['Attrition_flag'] = hr['Attrition'].astype(int)
else:
    # Try similar column names
    print("No 'Attrition' column found; adapt mapping accordingly.")

# Correlation with numeric features
numeric_cols = hr.select_dtypes(include=[np.number]).columns.tolist()
if 'Attrition_flag' in hr.columns:
    corr = hr[numeric_cols + ['Attrition_flag']].corr()['Attrition_flag'].sort_values(ascending=False)
    print("Top correlations with Attrition (numeric):")
    print(corr.head(10))

    # Visualize top drivers
    top_drivers = corr.abs().sort_values(ascending=False).index[1:6]  # top 5 excluding Attrition_flag itself
    for col in top_drivers:
        plt.figure(figsize=(6,3))
        sns.boxplot(x='Attrition_flag', y=col, data=hr)
        plt.title(f'{col} vs Attrition')
        plt.xlabel('Attrition (0 = No, 1 = Yes)')
        plt.show()
else:
    print("'Attrition_flag' not created; unable to compute numeric correlations.")

# Part b) Are employees with more projects more likely to leave?
if 'NumProjects' in hr.columns or 'NoOfProjects' in hr.columns:
    proj_col = 'NumProjects' if 'NumProjects' in hr.columns else 'NoOfProjects'
    proj_stats = hr.groupby(proj_col)['Attrition_flag'].mean().reset_index().rename(columns={'Attrition_flag':'attrition_rate'})
    print("\nAttrition rate by number of projects:")
    print(proj_stats)
    plt.figure(figsize=(8,4))
    plt.plot(proj_stats[proj_col], proj_stats['attrition_rate'], marker='o')
    plt.xlabel(proj_col)
    plt.ylabel('Attrition Rate')
    plt.title('Attrition rate vs Number of Projects')
    plt.grid(True)
    plt.show()
else:
    print("No project-count column found. Adapt column name accordingly.")
```
